```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

consumer_key = 'API_KEY'
consumer_secret = 'SECRET'
access_token = 'TOKEN'
access_secret = 'SECRET'
    
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth_handler=auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

### Quality

1. Completeness (missing values):
    - **twitter-archive-enhanced.csv**:`NaN` values in `in_reply_to_status_id`, `in_reply_to_user_id`, `retweeted_status_id`, `retweeted_status_timestamp` and `None` values in `name`,  `doggo`, `floofer`, `pupper`, `puppo`
    - `expanded_urls` has null values
    
    - **tweet_json.txt**: `NaN` values in `extended_entities`, `quoted_status`, `quoted_status_id`, `quoted_status_id_str`, `quoted_status_permalink`, `retweeted_status` and `None` values in `contributors`, `coordinates`, `geo`


2. Validity (naming conventions and grammatical errors): 

    - **twitter-archive-enhanced.csv**: single letters in dog's `name`, and `name` beginning with lower case. 
    - **image-predictions.tsv**: `p1, p2, p3` have both upper and lower case strings. `p1, p2, p3` have `_` between name, and columns names need to be renamed to show prediction outcome to be more readable


3. Accuracy (duplicates):
    - **twitter-archive-enhanced.csv**: duplicate tweets checking retweet column if they are empty or text starts with `RT@` including duplicate jpg image url's 
 
4. Consistency (metrics): 
    -  **twitter-archive-enhanced.csv**: `rating_numerator` column should be a float while incorrect values appear and not properly extracted from tweet text, and `rating_denominator` should be a float.

### Tidiness


- Column headers are values, not variable names.
- Multiple variables are stored in one column.
- Variables are stored in both rows and columns.
- Multiple types of observational units are stored in the same table.
- A single observational unit is stored in multiple tables.

1. all three `json`, `tsv`, `txt` DataFrames need to be merged to combine only important data joining on `tweet_id`
2. `tweet_json.txt`: remame `id_str` to `tweet_id` for merging dataframes
3. we should merge all columns `doggo, floofer, pupper and puppo`, into one column called dog_stage, as they are all aspects of the stage of a dog

## Part III: Clean

#### create copies of our df's to ensure we have backups

```python
# using .copy() we ensure we have a copy of each df 
archive_clean = df_archive.copy()
json_clean = df_json.copy()
tsv_clean = df_tsv.copy()
```

```python
# .columms allows us to see our df columns 
archive_clean.columns
```

```python
json_clean.columns
```

```python
tsv_clean.columns
```

#### drop columns, drop null values, convert to timestamp, ignore lower case names

```python
# .drop() allows us to drop columns in archive_clean not necessary and that were also missing values
archive_clean.drop(columns=['in_reply_to_status_id', 'in_reply_to_user_id', 'timestamp', 'retweeted_status_id', 'retweeted_status_user_id','retweeted_status_timestamp', 'expanded_urls'], inplace=True)
```

```python
# .drop() allows us to drop columns in json_clean that are not necessary 
json_clean.drop(columns=['contributors', 'coordinates', 'display_text_range', 'entities', 'extended_entities', 'favorited', 'geo', 'id_str', 'in_reply_to_screen_name', 'in_reply_to_status_id', 'in_reply_to_status_id_str', 'in_reply_to_user_id', 'in_reply_to_user_id_str', 'is_quote_status', 'lang', 'place', 'possibly_sensitive', 'possibly_sensitive_appealable','quoted_status', 'quoted_status_id', 'quoted_status_id_str', 'quoted_status_permalink', 'retweeted', 'retweeted_status', 'source', 'truncated', 'user'], inplace=True)
```

```python
# .drop() from tsv_clean our confidence intervals because we already have the outcomes as a boolean T/F
tsv_clean.drop(columns=['img_num', 'p1_conf', 'p2_conf', 'p3_conf'], inplace=True)
```

#### handle merging all dog stages to one column 

```python
# we replace all "None" values with an empty space using pandas .replace()
archive_clean[['doggo', 'floofer', 'pupper', 'puppo']] = archive_clean[['doggo', 'floofer', 'pupper', 'puppo']].replace('None', ' ')

# we concat each column after .replace() method into a new column called "dog_stage"
archive_clean['dog_stage'] = archive_clean['doggo']  + ' '  + archive_clean['floofer']  + ' ' +  archive_clean['pupper'] + ' '  +  archive_clean['puppo']

# we can now successfully .drop() our old stages columns and update our archive df
archive_clean = archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1)
```

#### remove duplicate tweets that start with "RT @"

```python
# let's get a sum of all tweets with the string, "RT @" and define the size of this issue
json_clean.full_text.str.contains("RT @").sum()
```

```python
# let's explicitly define our string into a variable called "searchfor"
searchfor = ['RT @']

# using pandas .contains() method, we check if "searchfor" is in our tweet's "full_text"  
# then using panda's boolean indexing, we are able to drop columns that contain the "searchfor" string
json_clean = json_clean[~json_clean.full_text.str.contains('|'.join(searchfor))]
```

```python
# let's test and recheck our sum of all tweets with the string, "RT @" which should now be 0 
json_clean.full_text.str.contains("RT @").sum()
```

#### remove lower case names

```python
# let's get a sum of all dog names that have a first letter as lower case to define our issue
archive_clean.name.str[0].str.islower().sum()
```

```python
# using the same method as above with boolean indexing, here we now check if the first letter of 
# our string as "str[0]" is lower case with islower()
archive_clean = archive_clean[~archive_clean['name'].str[0].str.islower()]
```

```python
# let's test and check the sum of all tweets with lowercase, which should now be 0
archive_clean.name.str[0].str.islower().sum()
```

#### extract ratings from tweet text and apply float

```python
# using pandas extract with regex we can find digits represented as a fraction
# where "\d" looks for digits within our tweets text
# we then set our rating variable results to the numerator/denominator columns
rating = archive_clean.text.str.extract('((?:\d+\.)?\d+)\/(\d+)', expand=True)
rating.columns = ['rating_numerator', 'rating_denominator']
```

```python
# let's test our code with value_counts() for our denominators
# it looks like we have many "10"'s
archive_clean.rating_denominator.value_counts()
```

```python
# let's test our numerators with value_counts() 
# it looks like we have many different numbers across the board!
archive_clean.rating_numerator.value_counts()
```

```python
# we must convert our new ratings from strings to floats
str_to_float = archive_clean[['rating_numerator', 'rating_denominator']]

# we iterate over our ratings colummns and use the pandas to_numeric() method 
# to convert values to int64's then we use apply() method to convert to float
for col in str_to_float:
    archive_clean[col] = pd.to_numeric(archive_clean[col], errors='coerce').apply(float)   # cite 11
```

```python
# let's test our datatypes, great, we see float!
archive_clean.rating_numerator.dtype
```

```python
# let's test our datatypes, great, we see float!
archive_clean.rating_denominator.dtype
```

```python
# we see the shape of our df's after all of these changes!
archive_clean.shape, json_clean.shape, tsv_clean.shape
```

#### convert created at columns to timestamps

```python
# let's make sure we have an accurate timestamp for our tweets
# we use pandas datetime to convert these timestamps to actual datetime
json_clean['created_at'] = pd.to_datetime(json_clean['created_at'])
```

```python
# let's use lambda to iterate over our datetime to correctly format each value
# we will add %I:%M %p to get hour, minute and either AM or PM! 
json_clean['created_at'] = json_clean['created_at'].map(lambda x: x.strftime('%Y%m%d%' + ' ' + '%I:%M %p') if x else "") # cite 2
```

#### drop text column as we already have tweets from json_clean

```python
# we can drop the tweet text from archive_clean since we already have this in json_clean!
archive_clean.drop(columns=['text'], inplace=True)
```

#### rename columns

```python
# let's rename our id to tweet_id so we can merge df's with this unique key
# let's also rename our created_at to timestamp to be more clear
json_clean = json_clean.rename(columns={"id": "tweet_id", "created_at": "timestamp"})
```

```python
# let's rename our ambiguous name column to dog_name
archive_clean = archive_clean.rename(columns={"name": "dog_name"})
```

```python
# let's rename our image predictions columns to a more readable name
# we are clearly stating, 1st, 2nd, and 3rd dog prediction outcomes
tsv_clean = tsv_clean.rename(columns={"p1": "first_dog_prediction", "p1_dog": "first_prediction_outcome", \
                         "p2": "second_dog_prediction", "p2_dog": "second_prediction_outcome", \
                          "p3": "third_dog_prediction", "p3_dog": "third_prediction_outcome"})
```

#### string function replace and title

```python
# each column had character "_" in between our dog names
# using pandas .replace() we can replace those values with a space instead! 
# finally, we can use pandas title() method to make sure all names are title case!
tsv_clean['first_dog_prediction'] = tsv_clean['first_dog_prediction'].astype(str).replace(to_replace ='_', value = ' ', regex = True) 
tsv_clean['first_dog_prediction'] = tsv_clean['first_dog_prediction'].astype(str).str.title()

tsv_clean['second_dog_prediction'] = tsv_clean['second_dog_prediction'].astype(str).replace(to_replace ='_', value = ' ', regex = True) 
tsv_clean['second_dog_prediction'] = tsv_clean['second_dog_prediction'].astype(str).str.title()

tsv_clean['third_dog_prediction'] = tsv_clean['third_dog_prediction'].astype(str).replace(to_replace ='_', value = ' ', regex = True) 
tsv_clean['third_dog_prediction'] = tsv_clean['third_dog_prediction'].astype(str).str.title()
```

```python
# let's check that all tweet_id's are the same datatype 
# this ensure we can properly merge on this unique key! 
tsv_clean.tweet_id.dtype, json_clean.tweet_id.dtype, archive_clean.tweet_id.dtype # check join key same type 
```

#### Merge DataFrames

```python
# here we can see all columns names after our cleaning!
json_clean.columns, tsv_clean.columns, archive_clean.columns
```

```python
# using functools and reduce we put all df's into a list, and into the variable dfs
# using reduce(), lambda() and merge() we can merge all df's "on" our "tweet_id" 
from functools import reduce
dfs = [json_clean, tsv_clean, archive_clean]
merged = reduce(lambda left,right: pd.merge(left,right,on='tweet_id'), dfs) # cite 3
```

```python
# testing our merged df with sample, and it looks great!
merged.sample(1)
```

```python
merged.columns
```

```python
# let's reorder our columns so everything is more clear when reading L-R
master = merged.reindex(columns=['timestamp', 'tweet_id', 'full_text', 'jpg_url', 'source', 'dog_name', 'dog_stage',
        'favorite_count', 'retweet_count', 'rating_numerator', 'rating_denominator',
       'first_dog_prediction', 'first_prediction_outcome','second_dog_prediction', 'second_prediction_outcome',
       'third_dog_prediction', 'third_prediction_outcome'])
```
